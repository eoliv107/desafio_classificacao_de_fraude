# Desafio de detecção de fraude-Nvidia

A base de dados que será utilizada para o desenvolvimento deste desafio contém aproximadamente 6,3 milhões de transações. Os dados contidos nesta base de dados são simulados e compreendem 30 dias. Estas transações podem ser de diferentes tipo e foram executadas via mobile devices. As features contidas nesta base de dados são:<br>

<b>- step - Passo:</b> representa o total de horas transcorrido desde o início da simulação. Esta feature vai variar entre 1 e 744 (30 dias);<br>

<b>- type - Tipo:</b> tipo de transação (depósito, saque, débito, pagamento e transferência);<br>

<b>- amount - Quantia:</b> total que foi transacionado;<br>

<b>- nameOrig - ClienteOrigem:</b> cliente que iniciou a transação<br>

<b>- oldbalanceOrg - SaldoInicialOrigem:</b> saldo da conta de origem antes da transação;<br>

<b>- newbalanceOrig - SaldoFinalOrigem:</b> saldo da conta de origem após a transação;<br>

<b>- nameDest - ClienteDestino:</b> cliente de destino da transação;<br>

<b>- oldbalanceDest - SaldoInicialDestino:</b> saldo da conta de destino antes da transação;<br>

<b>- newbalanceDest - SaldoFinalDestino:</b> saldo da conta de destino após a transação;<br>

<b>- isFraud - ÉFraude:</b> flag que define se a transação é fraudulenta ou não. Nesta simulação o objetivo da fraude é assumir a conta do usuário, esvaziá-la transferindo para outra conta e então sacando o dinheiro;<br>

<b>- isFlaggedFraud - SinalizadaComoFraude:</b> automaticamente marcadas pelo banco como fraude por tentarem transferir mais de 200.000 em uma única transação.<br>

```python
#Definindo algumas funções uteis para utilização posterior
#Função para calcular o VIF e retornar um dataframe com os valores para cada variável
def calcular_vif(df):

    vif = pd.DataFrame()
    vif['variaveis'] = df.columns
    vif['VIF'] = [variance_inflation_factor(df.values, i) for i in range(len(df.columns))]

    return(vif)

# Função de métricas
def metrics_report(y, y_pred, dataset_name='valid'):
    mean_y = np.mean(y)
    mean_pred = np.mean(y_pred)

    rmse = round(math.sqrt(mean_squared_error(y, y_pred)), 2)
    mae = mean_absolute_error(y, y_pred)
    r2 = r2_score(y, y_pred)
    mape = mean_absolute_percentage_error(y, y_pred)

    report = pd.DataFrame({
    'metric': ['RMSE', 'MAE', 'R^2', 'MAPE', 'Avg. target', 'Avg. Prediction'],
    dataset_name: [rmse, mae, r2, mape, mean_y, mean_pred]
    })

    report[dataset_name] = report[dataset_name].round(2)
    return report
```

```python
#Importando o dataset e visualizando as 5 primeiras linhas
dataset = pd.read_csv('data/fraud_detection_dataset.csv').clean_names()
dataset.head()
```

```python
#Vamos criar feature engineering considerando a quantidade de transações de cada name de origem e destino. Depois 
#faremos um índice
dic_origem = {}
for index, valor in zip(dataset.nameorig.value_counts().index, dataset.nameorig.value_counts().values):
    dic_origem[index] = valor
dic_destino = {}
for index, valor in zip(dataset.namedest.value_counts().index, dataset.namedest.value_counts().values):
    dic_destino[index] = valor

dataset['quantorigem'] = dataset.nameorig.map(dic_origem)
dataset['quantdestino'] = dataset.namedest.map(dic_destino)
dataset['faixaquantdestino'] = pd.cut(dataset.quantdestino, bins= [0, 31, 61, 91, dataset.quantdestino.max()],\
                                                            labels = ['0-30','31-60','61-90','91-acima'])
```

```python
#vamos transformar cada variável de acordo com seu tipo de dado
for valor in categorica:
    dataset[valor] = dataset[valor].astype('category')
for valor in numerica:
    dataset[valor] = dataset[valor].astype('float64')
for valor in objetivo:
    dataset[valor] = dataset[valor].astype('int64')
```

```python
# Vamos começar combinando max_depth/min_child_weight para encontrarmos o melhor resultado
gridsearch_params = [(max_depth, min_child_weight) for max_depth in range(9,12) for min_child_weight in range(5,8)]
# Definindo o melhor parametro inicial e MAE
min_mae = float("Inf")
best_params = None
num_boost_round = 999
for max_depth, min_child_weight in gridsearch_params:
    print("max_depth={}, min_child_weight={}".format(
                             max_depth,
                             min_child_weight))
    # Atualizando nossos parâmetros
    params['max_depth'] = max_depth
    params['min_child_weight'] = min_child_weight
    # rodar o xboost cv
    cv_results = cv(
        params,
        dtrain,
        num_boost_round=num_boost_round,
        seed=42,
        nfold=5,
        metrics={'mae'},
        early_stopping_rounds=10
    )
    # Atualizar o melhor MAE
    mean_mae = cv_results['test-mae-mean'].min()
    boost_rounds = cv_results['test-mae-mean'].argmin()
    print("\tMAE {} para {} iterações".format(mean_mae, boost_rounds))
    if mean_mae < min_mae:
        min_mae = mean_mae
        best_params = (max_depth,min_child_weight)
print("Best params: {}, {}, MAE: {}".format(best_params[0], best_params[1], min_mae))
```

```python
#COntinuanto a tunagem agora para o combinado de subsample/colsample para encontrarmos o melhor resultado
gridsearch_params = [(subsample, colsample) for subsample in \
                     [i/10. for i in range(7,11)]for colsample in [i/10. for i in range(7,11)]]
# Definindo o melhor parametro inicial e MAE
min_mae = float("Inf")
best_params = None
# Vamos iniciar pelos maiores valores e depois desceremos para o menor valor
for subsample, colsample in reversed(gridsearch_params):
    print("subsample={}, colsample={}".format(
                             subsample,
                             colsample))
    # Vamos atualizar os parametros
    params['subsample'] = subsample
    params['colsample_bytree'] = colsample
    # rodar o xboost cv
    cv_results = cv(
        params,
        dtrain,
        num_boost_round=num_boost_round,
        seed=42,
        nfold=5,
        metrics={'mae'},
        early_stopping_rounds=10
    )
    # Atualizar o melhor MAE
    mean_mae = cv_results['test-mae-mean'].min()
    boost_rounds = cv_results['test-mae-mean'].argmin()
    print("\tMAE {} para {} iterações".format(mean_mae, boost_rounds))
    if mean_mae < min_mae:
        min_mae = mean_mae
        best_params = (subsample,colsample)
print("Best params: {}, {}, MAE: {}".format(best_params[0], best_params[1], min_mae))
```

```python
#Continuanto a tunagem agora para o eta para encontrarmos o melhor resultado
#Vamos adicionar um contador de segundos

min_mae = float("Inf")
best_params = None
for eta in [.3, .2, .1, .05, .01, .005]:
    print("eta={}".format(eta))
    # vamos atualizar nossos parametros
    params['eta'] = eta
    # rodar o xboost cv
    cv_results = cv(params,
            dtrain,
            num_boost_round=num_boost_round,
            seed=42,
            nfold=5,
            metrics=['mae'],
            early_stopping_rounds=10)
    # Atualizar o melhor MAE
    mean_mae = cv_results['test-mae-mean'].min()
    boost_rounds = cv_results['test-mae-mean'].argmin()
    print("\tMAE {} para {} iterações\n".format(mean_mae, boost_rounds))
    if mean_mae < min_mae:
        min_mae = mean_mae
        best_params = eta
print("Best params: {}, MAE: {}".format(best_params, min_mae))
```
