# Desafio de detecção de fraude-Nvidia

A base de dados que será utilizada para o desenvolvimento deste desafio contém aproximadamente 6,3 milhões de transações. Os dados contidos nesta base de dados são simulados e compreendem 30 dias. Estas transações podem ser de diferentes tipo e foram executadas via mobile devices. As features contidas nesta base de dados são:<br>

<b>- step - Passo:</b> representa o total de horas transcorrido desde o início da simulação. Esta feature vai variar entre 1 e 744 (30 dias);<br>

<b>- type - Tipo:</b> tipo de transação (depósito, saque, débito, pagamento e transferência);<br>

<b>- amount - Quantia:</b> total que foi transacionado;<br>

<b>- nameOrig - ClienteOrigem:</b> cliente que iniciou a transação<br>

<b>- oldbalanceOrg - SaldoInicialOrigem:</b> saldo da conta de origem antes da transação;<br>

<b>- newbalanceOrig - SaldoFinalOrigem:</b> saldo da conta de origem após a transação;<br>

<b>- nameDest - ClienteDestino:</b> cliente de destino da transação;<br>

<b>- oldbalanceDest - SaldoInicialDestino:</b> saldo da conta de destino antes da transação;<br>

<b>- newbalanceDest - SaldoFinalDestino:</b> saldo da conta de destino após a transação;<br>

<b>- isFraud - ÉFraude:</b> flag que define se a transação é fraudulenta ou não. Nesta simulação o objetivo da fraude é assumir a conta do usuário, esvaziá-la transferindo para outra conta e então sacando o dinheiro;<br>

<b>- isFlaggedFraud - SinalizadaComoFraude:</b> automaticamente marcadas pelo banco como fraude por tentarem transferir mais de 200.000 em uma única transação.<br>

```python
#Definindo algumas funções uteis para utilização posterior
#Função para calcular o VIF e retornar um dataframe com os valores para cada variável
def calcular_vif(df):

    vif = pd.DataFrame()
    vif['variaveis'] = df.columns
    vif['VIF'] = [variance_inflation_factor(df.values, i) for i in range(len(df.columns))]

    return(vif)

# Função de métricas
def metrics_report(y, y_pred, dataset_name='valid'):
    mean_y = np.mean(y)
    mean_pred = np.mean(y_pred)

    rmse = round(math.sqrt(mean_squared_error(y, y_pred)), 2)
    mae = mean_absolute_error(y, y_pred)
    r2 = r2_score(y, y_pred)
    mape = mean_absolute_percentage_error(y, y_pred)

    report = pd.DataFrame({
    'metric': ['RMSE', 'MAE', 'R^2', 'MAPE', 'Avg. target', 'Avg. Prediction'],
    dataset_name: [rmse, mae, r2, mape, mean_y, mean_pred]
    })

    report[dataset_name] = report[dataset_name].round(2)
    return report
def calculate_far_frr(y_proba, y_true, n_points=30, normalize=False, threshold=None):
    thr = np.linspace(0, 1, n_points)
    far = []
    frr = []
    for t in thr:
        far_pt = y_proba[y_true == 0] >= t
        frr_pt = y_proba[y_true == 1] <= t
        if normalize:
            far.append(far_pt.mean())
            frr.append(frr_pt.mean())
        else:
            far.append(far_pt.sum())
            frr.append(frr_pt.sum())
    if threshold is None:
        far = np.array(far)
        frr = np.array(frr)
        optm = thr[abs(far - frr).argmin()]
    else:
        optm = threshold
    return {
        'optm': optm,
        'plot': pd.DataFrame(
            index=thr,
            data={
                'far': far,
                'frr': frr,
            },
            dtype=float
        )
    }
```

```python
#Importando o dataset e visualizando as 5 primeiras linhas
dataset = pd.read_csv('data/fraud_detection_dataset.csv').clean_names()
dataset.head()
```

```python
#Vamos criar feature engineering considerando a quantidade de transações de cada name de origem e destino. Depois 
#faremos um índice
dic_origem = {}
for index, valor in zip(dataset.nameorig.value_counts().index, dataset.nameorig.value_counts().values):
    dic_origem[index] = valor
dic_destino = {}
for index, valor in zip(dataset.namedest.value_counts().index, dataset.namedest.value_counts().values):
    dic_destino[index] = valor

dataset['quantorigem'] = dataset.nameorig.map(dic_origem)
dataset['quantdestino'] = dataset.namedest.map(dic_destino)
dataset['faixaquantdestino'] = pd.cut(dataset.quantdestino, bins= [0, 31, 61, 91, dataset.quantdestino.max()],\
                                                            labels = ['0-30','31-60','61-90','91-acima'])
```

```python
#Verificando qual o tipo de dados das variáveis do dataset
dataset.info()
```

```python
#Criar as variáveis para cada tipo de dados do dataset
categorica = ['type', 'nameorig', 'namedest', 'faixaquantdestino']
numerica = ['step', 'amount', 'oldbalanceorg', 'newbalanceorig', 'oldbalancedest', 'newbalancedest',
            'quantorigem', 'quantdestino']
objetivo = ['isfraud', 'isflaggedfraud']
```

```python
#vamos transformar cada variável de acordo com seu tipo de dado
for valor in categorica:
    dataset[valor] = dataset[valor].astype('category')
for valor in numerica:
    dataset[valor] = dataset[valor].astype('float64')
for valor in objetivo:
    dataset[valor] = dataset[valor].astype('int64')
```

```python
#Vamos analisar o tamanho do dataset
dataset.shape
```

```python
#Analisando valores nulos para cada variavel
dataset.isnull().mean()
```

```python
#Fazendo uma contagem de do conteúdo de fraude e não fraude
pd.DataFrame(dataset.isfraud.value_counts().values, index=['Não é fraude','É Fraude'], columns=['Quantidade'], )
```

```python
#Contar os valores de fraude e não fraude pelo tipo de transação
dataset.groupby(['isfraud','type'])[['amount']].count()
```

```python
#Pre-processamento dos datasets de treino, teste e validação
X = dataset.drop(columns=('isfraud'))
y = dataset.isfraud
```

```python
# Pré-processamento das variáveis categóricas (categorical encoders)
X[categorica] = TargetEncoder().fit_transform(X[categorica], y)
X.head()
```

```python
#Correlação entre os features mais importantes para a classificação
plt.figure(figsize=(15,10))
ax = sns.heatmap(dataset.corr(), annot=True, cmap = 'PuRd')
plt.title('Correlação das variáveis numéricas', fontsize = 14, loc = 'left')
```

Na matriz de correlação vemos um alto indece de correlação entre os balanços de antes de depois de quem faz e quem recebe a transação, sendo assim vamos optar por retirar um deles no momento da classificação

```python
#Dividindo dataset em treino e teste
X_train, X_test, y_train, y_test = train_test_split(X,y,train_size=0.7, random_state=42)
```

```python
#Vamos calcular o VIF para as variáveis explicativas
calcular_vif(X)
```

Como podemos ver, oldbalanceorg, newbalanceorig, oldbalancedest e newbalancedest possuem valores de VIF muito elevados, indicando que as variáveis estão altamente correlacionadas. Isso é esperado porque as contas da origem e destino antes e depois se correlacionam entre si. Portanto, considerar essas duas características juntas leva a um modelo com alta multicolinearidade.

### 1 - Utilizando o LogisticRegression

```python
#Fitando o Modelo utilizando o logistic regression
lr = LogisticRegression().fit(X_train, y_train)
y_pred_lr = lr.predict(X_test)
# Vendo a importancia
coef = lr.coef_[0]
features = X_train.columns
coef,names = zip(*sorted(zip(coef,features)))
plt.barh(range(len(features)), coef, align='center')
plt.yticks(range(len(features)), features)
plt.show()
```

```python
#gerando e plotando a matriz de confusão
plot_confusion_matrix(lr, X_test, y_test, cmap='YlOrRd')
```

```python
#Imprimíndo as métricas do modelo
print(f'Acurácia: {accuracy_score(y_test, y_pred_lr)}\n\
Precisão: {precision_score(y_test, y_pred_lr)}\n\
Recall: {recall_score(y_test, y_pred_lr)}')
```

```python
#Plotando a curva ROC
plot_roc_curve(lr, X_test, y_test)
```

### Gerando um sumário de todas as regressões juntamente com a sobreposição das curvas ROC de todas as predições

```python
#Gerando todos os gráficos para fazermos um comparativo entre os modelos
classifiers = [lr, svm, knn, nb, dtree, clf, rfc, adc, xgb_reg]
plt.figure(figsize=(15,10))
ax = plt.gca()
for i in classifiers:
    plot_roc_curve(i, X_test, y_test, ax=ax)
```

```python
#Plotando os resultados de todos os testes
df_aux = pd.DataFrame(columns = ['Classificador','Acuracia','Precisao','Recall', 'F1-Score'], index=range(len(classifiers)))
for i,value in enumerate(classifiers):
    y_pred = value.predict(X_test)
    df_aux.iloc[i]['Classificador'] = type(value).__name__
    df_aux.iloc[i]['Acuracia'] = round(accuracy_score(y_test, y_pred), 4)
    df_aux.iloc[i]['Precisao'] = round(precision_score(y_test, y_pred), 4)
    df_aux.iloc[i]['Recall'] = round(recall_score(y_test, y_pred), 4)
    df_aux.iloc[i]['F1-Score'] = 2*(((precision_score(y_test, y_pred))*(recall_score(y_test, y_pred)))/\
                                    ((precision_score(y_test, y_pred))+(recall_score(y_test, y_pred))))
```
